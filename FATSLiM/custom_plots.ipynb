```python
import numpy as np
import gromacs as gro
import matplotlib.pyplot as plt
```

```python
xvg = gro.fileformats.XVG('thickness.xvg')
xvg
```

```python
print(xvg.error) # kJ/mol
print(xvg.tc/1000) # put correlation time in ns
```

```python
data = xvg.array
x = data[0,:]
y = data[1,:]
x = x/1000 + 15
plt.plot(x,y,c='navy',label='membrane thickness')
plt.xlabel('Time (ns)')
plt.ylabel('Thickness (nm)')
plt.title('Thickness at Equilibrium');
```
